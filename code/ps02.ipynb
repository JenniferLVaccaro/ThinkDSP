2) Write a class called SawtoothChirp that extends Chirp and overrides evaluate to generate a sawtooth waveform with frequency that increases (or decreases) linearly. Plot the spectrogram of this signal. The effect of aliasing should be visible, and if you listen carefully, you can hear it.

```python
import math
PI2 = 2 * math.pi

class SawtoothChirp(thinkdsp.Chirp):
    """Represents a sawtooth signal with varying frequency."""
    def _evaluate(self, ts, freqs):
        """Helper function that evaluates the signal."""
        dts = numpy.diff(ts)
        dps = PI2 * freqs * dts
        phases = numpy.cumsum(dps)
        phases = numpy.insert(phases, 0, 0)
        cycles = phases / PI2
        frac, _ = numpy.modf(cycles)
        ys = thinkdsp.normalize(thinkdsp.unbias(frac), self.amp)
        return ys
```

```python
signal = SawtoothChirp(start=220, end=880)
wave = signal.make_wave(duration=3, framerate=10000)
wave.apodize()
wave.make_audio()
```

```python
spectrogram = wave.make_spectrogram(seg_length=512)
spectrogram.plot(high=5000)
thinkplot.config(xlabel='time(s)', ylabel='frequency (Hz)', legend=False)
```

The spectrogram for the Sawtooth Wave Chirp shows the same drop off of frequencies at each domininent/fundamental frequency, as expected. There appears to be a series of faint crossing lines, somewhat like an argyle sweater, in the spectrogram. This may be the visual effect of aliasing. The audible effect can be heard in the background the the recording; it sounds like the rushing sound which results from squeegying shampoo out of one's ear, or like a higher frequency version of listening to the ocean in a conch shell.

3)  In musical terminology, a “glissando” is a note that slides from one pitch to another, so it is similar to a chirp. A trombone player can play a glissando by extending the trombone slide while blowing continuously. As the slide extends, the total length of the tube gets longer, and the resulting pitch is inversely proportional to length. Assuming that the player moves the slide at a constant speed, how does frequency vary with time? Is a trombone glissando more like a linear or exponential chirp? Write a function that simulates a trombone glissando from C3 up to F3 and back down to C3. C3 is 262 Hz; F3 is 349 Hz.

```python
class TromboneGliss(thinkdsp.Chirp):
    """Represents a trombone-like signal with varying frequency."""
    def evaluate(self, ts):
        """Evaluates the signal at the given times."""
        l1, l2 = 1.0 / self.start, 1.0 / self.end
        lengths = numpy.linspace(l1, l2, len(ts)-1)
        freqs = 1 / lengths
        return self._evaluate(ts, freqs)
```

```python
low = 262 #This is the frequency of a C3
high = 349 #This is the frequency of a F3
trom1 = TromboneGliss(low, high)
trom2 = TromboneGliss(high, low)
wave1 = trom1.make_wave(duration=2)|trom2.make_wave(duration=1)
wave1.apodize()
wave1.make_audio()
```

```python
tromspectrogram = wave1.make_spectrogram(seg_length=512)
tromspectrogram.plot(high=35)
thinkplot.config(xlabel='time(s)', ylabel='frequency (Hz)', legend=False)
```

As a former member of marching band, I am fairly confident that a trombone glissando with a linear increase in slide length changes in pitch linearly. This is the same as changing in frequency exponentially, because the fundamental frequency will be inversely proportional to slide length. The spectrogram is a little difficult to see clearly, but the rate of frequency increase with respect to time seems to increase at higher frequencies, suggesting an exponential.

6) In a white noise signal, the mixture of frequencies changes over time. In the long run, we expect the power at all frequencies to be equal, but in any sample, the power at each frequency is random.
To estimate the long-term average power at each frequency, we can break a long signal into segments, compute the power spectral density for each segment, and then compute the average across the segments. You can read more about this algorithm at http://en.wikipedia.org/wiki/Bartlett's_method.

```python
N = 350 #for any N point data segment dummywave
K = 50 #total number of data segments, integer factor of N
M = N/K #length of each data segment
spectrumsum = 0
bigdummywave = thinkdsp.UncorrelatedGaussianNoise().make_wave()

for i in range(1,K):
        dummywave = thinkdsp.UncorrelatedGaussianNoise().make_wave()
        dummywave.segment(i*M,M)
        dummyspectrum = dummywave.make_spectrum()
        spectrumsum += dummyspectrum.hs**2/M
spectrum(hs=spectrumsum)
spectrum.plot(high=2000)
thinkplot.config(xlabel='frequency (Hz)', ylabel='amplitude', legend=False)
```

According to the wikipedia page given in the article, the Bartlett method seems to take a data set (such as wave "dummywave") of length N and break it up into K equal sized segments of length M. M and K are both integer factors of N, so that there is no overlap of the segments. Then, for each segment, you take the DFT, square that result, and divide it by M. The average of these over the K segments will be the array of power measurements vs. frequency.
