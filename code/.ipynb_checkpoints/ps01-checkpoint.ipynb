Importing the important imports!


```python
from __future__ import print_function, division

import thinkdsp
import thinkplot
from IPython.display import Audio
from IPython.html.widgets import interact, fixed
from IPython.html import widgets
```

```python
wave = thinkdsp.read_wave('203726__genelythgow__4-bell-piano-hhits.wav')
wave.make_audio()
```

Q1: This is a recording of a bell piano.

```python
segment = wave.segment(start=8.5, duration=0.0015)
segment.normalize()
segment.apodize()
segment.plot()
```

```python
spectrum = segment.make_spectrum()
spectrum.plot()
```

Q3: The spectrum does not conclusively show a harmonic structure, because the multiples of the lowest frequency are not obviously represented.

```python
spectrumLP = wave.segment(8.5,1).make_spectrum()
spectrumLP.low_pass(10000)
spectrumLP.make_wave().make_audio()
```

Q4: A low pass filter resulted in a less rich timbre. The high pass did as well, and I believe that I could still hear the dominent frequency. A bandstop filter resulted in a somewhat thinner sound, but had a less distinct change on the timbre than I expected.

Q5: I synthesized a wave using several frequency components.

```python
spectrumHP = wave.segment(8.5,1).make_spectrum()
spectrumHP.high_pass(10000)
spectrumHP.make_wave().make_audio()
```

```python
spectrumBS = wave.segment(8.5,1).make_spectrum()
spectrumBS.band_stop(5000,15000)
spectrumBS.make_wave().make_audio()
```
